```python
def get_laptop_data():
    '''
    Basic function to full laptop data from the csv file.
    '''
    df = pd.read_csv('Cleaned_Laptop_data.csv')
    
    return df
```

```python
def clean_data():
    '''
    This function will use the above function to get the laptop data, as well as clean the data in a way that makes it easier to work with and make sense.
    '''
    # Function to get the basic df from csv file
    df = get_laptop_data()
    # Converting the prices in the dataset from Indian Rupees to USD
    df['latest_price'] = df['latest_price'] * 0.01220
    df['old_price'] = df['old_price'] * 0.01220
    # Lowercasing the column names and specifically the brand column to simplify working with
    df.columns = df.columns.str.lower()
    df['brand'] = df['brand'].str.lower()
    # Changing the yes's and no's to numerical values
    df = df.replace('Yes', 1)
    df = df.replace('No', 0)
    # Renaming reviews column to num_reviews
    df = df.rename(columns={'reviews':'num_reviews'})
    # Changing the missing values from the os column to windows
    df['os'] = df['os'].replace('Missing', 'Windows')
    # Renaming the processor generation column
    df = df.rename(columns={'processor_gnrtn':'processor_gen'})
    # Making a list of columns with the object data type
    lst = ['brand', 'model', 'processor_brand', 'processor_name', 'processor_gen', 'ram_gb', 'ram_type', 'os', 'weight', 'display_size']
    # Removing all null values (only 25) across the whole df
    df = df[~df['ram_gb'].isna()]
    # Small loop to lowercase the entries for each column in the above list
    for col in df[lst]:
    
        df[col] = df[col].str.lower()
        
    # The next several lines are for the purpose of replacing the 0s in the old_price column with the same entry in the latest_price
    # This subsets the data into a df with all of the rows where the old_price is 0
    temp = df[df['old_price'] == 0.0000]
    # This simply adds the two columns together for each for so that they're the same
    temp['old_price'] = temp['latest_price'] + temp['old_price']
    # Making each row where the old_price is less than 1 into a null
    df['old_price'] = df['old_price'].drop(df[df['old_price'] < 1].index)
    # Dropping each row where there's a null in old_price
    # Alternatively, instead of this and the above, could simply take this syntax and filter out each entry where the old_price is less than 1 and some a line of code. :shrug:
    df = df[~df['old_price'].isna()] 
    # Concatenating the two dataframes
    df = pd.concat([df, temp])
    # Resetting the indices, the drop parameter here drops the old index, because otherwise it would be made into a new (and useless) column.
    # inplace=True simply means I don't need to reassign the variable
    df.reset_index(drop=True,inplace=True)
    # Creating an average of the old and latest prices
    df['working_average'] = (df['latest_price'] + df['old_price']) / 2
    
    # ...returns the dataframe
    return df
```
